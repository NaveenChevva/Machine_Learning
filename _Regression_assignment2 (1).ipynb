```python
df["brand_name"]=df.groupby("brand_name")["processor_brand"].transform(lambda x:x.fillna(x.mode()))  # Fill missing values with mode
df["brand_name"]=df.groupby("brand_name")["os"].transform(lambda x:x.fillna(x.mode()))
df.drop("brand_name",axis=1,inplace=True)
```

```python
for i in df.columns:
    if df[i].dtype != 'object':  # Check if the column is numeric
        median = df[i].median()
        df[i] = df[i].fillna(median)
    elif df[i].dtype == 'object':  # Handle object (string) columns
        mode = df[i].mode()[0]  # Get the most frequent value
        df[i] = df[i].fillna(mode)  # Fill missing values with mode
```

```python
a=1
plt.figure(figsize=(15,30))
for i in df[["price","processor_brand","battery_capacity","ram_capacity","internal_memory","screen_size","num_rear_cameras","num_cores"]]:
    if df[i].dtype!="object":
        plt.subplot(9,2,a)
        sns.distplot(df[i])
        plt.subplot(9,2,a+1)
        sns.boxplot(df[i])
        a+=2
plt.show()
```

```python
def remove_outlier(col):
    sorted(col)
    q1,q3=np.quantile(col,[0.25,0.75])
    iqr=q3-q1
    ll=q1-1.5*(iqr)
    ul=q3+1.5*(iqr)
    return ll,ul
```

```python
for i in df[
    [
        "price",
        "processor_brand",
        "battery_capacity",
        "ram_capacity",
        "internal_memory",
        "screen_size",
        "num_cores"   
]
]:
    if df[i].dtype!="object":
        ll,ul=remove_outlier(df[i])
        df[i]=np.where(df[i]>ul,ul,df[i])
        df[i]=np.where(df[i]<ll,ll,df[i])
    
```

```python
df['has_nfc'] = df['has_nfc'].astype(int)
df['has_nfc'] = df['has_nfc'].replace({'False': 0, 'True': 1})
df['has_nfc'].value_counts()
```

```python
df['has_ir_blaster'] = df['has_ir_blaster'].astype(int)
df['has_ir_blaster'] = df['has_ir_blaster'].replace({'True':1, 'False':0})
```

```python
df['has_5g'] = df['has_5g'].astype(int)
df['has_5g'] = df['has_5g'].replace({'True':1, 'False':0})
```

```python
df["fast_charging_available"] = df['fast_charging_available'].astype(int)
df['fast_charging_available'] = df['fast_charging_available'].replace({'True':1, 'False':0})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
from math import sqrt
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
RFG = RandomForestRegressor(
    n_estimators=120,
    random_state=40,
    max_samples=1.0,
    max_features=0.4,
    max_depth=25
)
RFG.fit(X_train, y_train)
y_pred = RFG.predict(X_test)

print('R2 Score: ', r2_score(y_test, y_pred))
print('MAE : ', mean_absolute_error(y_test, y_pred))
print('MSE : ', mean_squared_error(y_test, y_pred))
print('RMSE : ', sqrt(mean_squared_error(y_test, y_pred)))
```
